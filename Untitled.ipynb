```python
m = ['o','^','v']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
for i, loc in enumerate(locations):
    file0 = loc + '/condAvePp' + '' + 'alpp1X1X1'
    data0 = np.loadtxt(file0)
    print(location)
    for j, a in enumerate(cases):
        file = loc + '/condAvePp' + a + 'alpp1X1X1'
        data = np.loadtxt(file)
        if i == 0:
            ax.scatter(data[:,0], data[:,1]/data0[:,1],marker=m[j],facecolors='none',edgecolors='k')
        else:
            ax.scatter(data[:,0], data[:,1]/data0[:,1],marker=m[j],color = 'k',edgecolors='k')        
ax.legend(['xx, <$\phi$>= 0.1','yy, <$\phi$>= 0.1','zz, <$\phi$>= 0.1',
           'xx, <$\phi$>= 0.3','yy, <$\phi$>= 0.3','zz, <$\phi$>= 0.3'],
         loc=1)
plt.xlabel('$\phi$',fontsize='large')
plt.ylabel('$\sigma_i/[1/3tr(\sigma)]$',fontsize='large')
plt.ylim(0.7, 1.6)
import matplotlib as mpl
mpl.rc("savefig", dpi=1000)
plt.savefig('xxyyzz.eps')
```
